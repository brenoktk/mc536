## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label=aspirin&match=exact&limit=10"

concept = "Aspirin"
match = "exact"
limit = "10"

concept = concept.replace(" ", "%20")

url = url_template.replace("{concept}", concept)\
                  .replace("{match}", match)\
                  .replace("{limit}", limit)

data = requests.get(url)
json_mi = data.json()
print(json.dumps(json_mi, indent=3))
```

Recuperando os termos a partir do código:

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = "D001241"

url = url_template.replace("{code}", code)

data = requests.get(url)
json_mid = data.json()
#print(json.dumps(json_mid, indent=3))
print("Termos da aspirina:")
for i in range(len(json_mid['terms'])):
    print(json_mid['terms'][i]['label'])
```

## REST API do OpenStreetMap para Query

```python
url_template = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Camp Nou"

url = url_template.replace("{format}", format)\
                  .replace("{query}", query)

data = requests.get(url)
campo_ref = data.json()
print(json.dumps(campo_ref, indent=3))
```

## Extraindo o Identificador OSM

```python
print(campo_ref[0]["osm_id"])
```

## REST API do OpenStreetMap para Elementos (Node, Way e Relation)

```python
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"
element_type = "way" # types: node, way, or relation
id = str(campo_ref[0]["osm_id"])
format = "json"

url = url_template.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

data = requests.get(url)
campo_way = data.json()
print(json.dumps(campo_way, indent=3))
```

## Extraindo o Identificador DBPedia

```python
print(campo_way["elements"][0]["tags"]["wikipedia"])
```

## REST API do DBPedia

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"
concept = campo_way["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                        .replace("en:", "")
format = "json"

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
campo_dbpedia = data.json()
print(json.dumps(campo_dbpedia, indent=3))
```
